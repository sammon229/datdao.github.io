# Identifying outliers



```python
def outliers_cal(df, column):
# Calculate z-scores for data points
    z_scores = (df[column] - np.mean(df[column])) / np.std(df[column])

# Identify outliers with z-scores greater than 2 or less than -2
    outliers = df[column][np.abs(z_scores) > 2]

# Print the indices of the outliers
    print("Indices of outliers:", outliers)
    
```

```python
outliers_cal(df, 'SalePrice')
```

```python
def identify_outliers(df, column):
    L = list()
    IQR = df[column].quantile(0.75)-df[column].quantile(0.25)
    Lowerfence_Rating = (df[column].quantile(0.25) - (1.5 * IQR))
    Upperfence_Rating = (df[column].quantile(0.75) + (1.5 * IQR))
    for x in df[column]:
        if x > Upperfence_Rating or x < Lowerfence_Rating :
            L.append(x)
    L.sort()
    print('Lowerfence_Rating is', Lowerfence_Rating)
    print('Lowerfence_Rating is', Upperfence_Rating)
    print('The existing outliers for', column, 'is', L)
```

```python
identify_outliers(df, 'GrLivArea')
identify_outliers(df, 'SalePrice')
identify_outliers(df, 'BedroomAbvGr')
```

To detect outliers, one can calculate the interquartile range and consider values beyond that range as outliers. Values below the lower fence and above the upper fence do not need to be considered. If necessary, outliers can be removed depending on the intended purpose.

# Covariance & Correlation


```python
cleansed_Data = df.dropna(subset=['GrLivArea', 'SalePrice'])
```

```python
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)
    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)
    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov
```

```python
GrLivArea, SalePrice = cleansed_Data['GrLivArea'], cleansed_Data['SalePrice']
Cov(GrLivArea, SalePrice)
```

def Corr(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)
    meanx, varx = np.mean(xs), np.var(xs)
    meany, vary = np.mean(ys), np.var(ys)
    corr = Cov(xs, ys, meanx, meany) / np.sqrt(varx * vary)
    return corr

#Pearsons Correlation coefficient
Corr(GrLivArea, SalePrice)

```python
def corr_coef(xs, ys):
    corr_coef = Cov(xs, ys) / (np.std(xs) * np.std(ys))
    return corr_coef
```

```python
# Correlation coefficient
corr_coef(GrLivArea, SalePrice)
```

The Pearson correlation coefficient is a measure of the strength and direction of the linear relationship between two variables. The Pearson's correlation coefficient between GrLivArea and SalePrice of 0.5893156056586915 indicates a moderately strong positive linear relationship between the two variables.A positive correlation means that as one variable increases, the other variable also tends to increase. In this case, as the living area (GrLivArea) of a house increases, the sale price of the house (SalePrice) also tends to increase.The magnitude of the correlation coefficient, which ranges from -1 to 1, indicates the strength of the relationship. A coefficient of 0 indicates no linear relationship, while a coefficient of 1 indicates a perfect positive linear relationship, and a coefficient of -1 indicates a perfect negative linear relationship.A coefficient of 0.5893156056586915 indicates a moderately strong positive linear relationship, which means that GrLivArea and SalePrice are positively correlated, but there may still be some unexplained variability in the data that is not captured by the linear relationship.

```python
from scipy.stats import spearmanr

spearman_corr, p_value = spearmanr(GrLivArea, SalePrice)
spearman_corr, p_value
```

A value of 0.6258949464633632 indicates a positive monotonic relationship between the two variables, where higher values of GrLivArea tend to be associated with higher values of SalePrice.The second value returned by the spearmanr function (p_value) is the associated p-value, which is a measure of the statistical significance of the correlation coefficient. In this case, the p-value is 2.260393e-317, which is very small (close to zero), indicating that the correlation coefficient is statistically significant and unlikely to have occurred by chance. This suggests that there is a real and significant relationship between the GrLivArea and SalePrice variables, based on their ranks.

# Hypothesis testing


```python
#Classical hypothesis testing on GrLivArea
stat, p_value = sp.stats.ttest_rel(df['GrLivArea'],df['SalePrice'])
print('p-value: ',p_value)
```

```python
from scipy.stats import norm

# Population mean
mu = 1500

# Sample mean and standard deviation
xbar = df['GrLivArea'].mean()
s = df['GrLivArea'].std(ddof=1)

# Sample size
n = len(df['GrLivArea'])

# Calculate the test statistic
z = (xbar - mu) / (s / np.sqrt(n))

# Calculate the p-value
p_value = 2 * norm.cdf(-np.abs(z))

# Print the results
print('Test statistic: ', z)
print('P-value: ', p_value)
```

The mean GrLivArea is equal to 1500 square feet. Our null hypothesis would be:
H0:  = 1500
And our alternative hypothesis would be:
Ha:   1500
The test statistic in this context is a measure of how much the sample data deviates from the null hypothesis, which is a hypothesis that assumes there is no significant difference between the two groups being compared. In this case, the test statistic is 0.026258210735064715.
The p-value is 0.9790513864104508, which is quite high. This indicates that the observed test statistic is not significant, and we cannot reject the null hypothesis. This means that there is no evidence to suggest that there is a significant difference between the two groups being compared.

```python
import numpy as np
import statsmodels.api as sm

# Select only numeric or continuous columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Remove response variable (SalePrice) from predictor variables
numeric_cols.remove('SalePrice')

# Convert columns to array
X = df[numeric_cols].to_numpy()
y = df['SalePrice'].to_numpy()

# Add constant term to predictor variables
X = sm.add_constant(X)

# Fit multiple linear regression model
model = sm.OLS(y, X).fit()

# Print summary of model statistics
print(model.summary())
```

R-squared: This is the coefficient of determination, which measures the proportion of variance in the response variable that is explained by the predictor variables. In this case, the R-squared value is 0.452, which means that the predictor variables explain 45.2% of the variance in the sale price.

Adj. R-squared: This is the adjusted R-squared value, which takes into account the number of predictor variables in the model. In this case, the adjusted R-squared value is 0.449.

F-statistic: This is the F-statistic for the overall significance of the model. It tests whether the predictor variables as a group have a significant effect on the response variable. The F-statistic value is 171.2, and the p-value for the F-test is 0.00, which means that the model is statistically significant.

Coefficients: These are the estimated regression coefficients for each predictor variable. The coefficient represents the change in the response variable (sale price) associated with a one-unit increase in the predictor variable, holding all other variables constant. The intercept (const) coefficient is -21010, which means that the predicted sale price when all predictor variables are equal to zero is $21,010. The other coefficients represent the change in the sale price associated with a one-unit increase in each predictor variable.

Standard errors: These are the standard errors of the regression coefficients, which measure the variability of the estimated coefficients.
The const coefficient has a t-value of -0.488 and a p-value of 0.625, indicating that it is not statistically significant at the conventional alpha level of 0.05. This means that the intercept term is not significantly different from zero, and the regression line is not significantly different from intersecting the y-axis at zero.
